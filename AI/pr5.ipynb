```python
import math
class xtrig:
    @staticmethod
    def cos(x):
        return math.cos(x)
    @staticmethod
    def sin(x):
        return math.sin(x)
    @staticmethod
    def tan(x):
        return math.tan(x)
    @staticmethod
    def arctan(x):
        return math.atan(x)
    @staticmethod
    def arcsin(x):
        return math.asin(x)
    @staticmethod
    def arccos(x):
        return math.acos(x)
    @staticmethod
    def deg2rad(x):
        return math.radians(x)

print(
    xtrig.arcsin(xtrig.cos(xtrig.deg2rad(90)))
)
```

# 1.2.4

```python
class xnode: 
    def __init__(self, data):
        self.left = self.right = None
        self.data = data
        
    def push(self,data):
        if data < self.data:
            if self.left:
                self.left.push(data)
            else:
                self.left=xnode(data)
        else:
            if data == self.data:
                return
            if self.right:
                self.right.push(data)
            else:
                self.right=xnode(data)
                
    def print(self,n=0):
        if self.right:
            self.right.print(n+5)
            
        print(' ' * n + str(self.data))
        
        if self.left:
            self.left.print(n+5)

MyTree = xnode(-2)
for i in range(1,20):
    MyTree.push(2^i % 17)
MyTree.print()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import tree
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn import metrics

dataset = pd.read_csv("wine.csv")
dataset.head()

x = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values

x_train, x_test, y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=1)
regressor = DecisionTreeRegressor(max_depth = 6)
regressor.fit(x_train,y_train)
tree.plot_tree(regressor)

y_pred = regressor.predict(x_test)

print('Средняя квадратичная ошибка:',metrics.mean_squared_error(y_test,y_pred))
print('Оценка:',np.round(metrics.r2_score(y_test,y_pred) * 100,2),'%')
```
