```python
import piplite as pl
await pl.install('seaborn')
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import numpy as np
from sklearn.cluster import KMeans

x = np.array(([
    [5,3],
    [10,15],
    [15,12],
    [24,10],
    [30,45],
    [85,70],
    [71,80], 
    [60,78],
    [55,52],
    [80,91]
]))

kmeans = KMeans(n_clusters=3)
kmeans.fit(x)
y_kmeans = kmeans.predict(x)
plt.scatter(x[:,0],x[:,1],c=y_kmeans,s=20,cmap='winter')
centers = kmeans.cluster_centers_
plt.scatter(centers[:,0],centers[:,1],c='blue',s=100,alpha=0.2)
plt.show()
```

```python
from sklearn.datasets import load_iris

data = load_iris()
x = data.get('data')

kmeans = KMeans(n_clusters = 5)
kmeans.fit(x)
y_kmeans = kmeans.predict(x)

display(x[:,0])

plt.scatter(x[:,0], x[:,1], c = y_kmeans, s = 20, cmap = 'spring')
centers = kmeans.cluster_centers_
plt.scatter(centers[:,0], centers[:,1], c = 'blue', s = 100, alpha = 0.2)
```
