# 4.3.2

```python
def func(x):
    return np.sqrt( 1 + np.power( np.e , np.sqrt(x) ) + np.cos( x**2 ) ) / abs( 1 - np.sin(x) ** 3 ) + np.log( abs(2*x) )

x = np.arange(0, 10)
y = np.vectorize(func)(x)

x1,y1 = x[0:5],y[0:5]
```

```python
plt.grid()
plt.plot(x1,y1)
```

# 4.3.3

```python
from scipy.integrate import simps
from numpy import trapz

def func(x):
    return abs( np.cos(x * np.power( np.e , np.cos(x) + np.log(x + 1) ) ) )

x = np.arange(0, 10)
y = np.vectorize(func)(x)

print("Площадь: ",trapz(y,x))

plt.grid()
plt.plot(x,y,c="black")
plt.fill_between(x,y,color="grey")
```

# 4.3.5

```python
import numpy as np

c,x,y = "+ 2 3".split() #input().split()
x = int(x)
y = int(y)

if c == "+":
    print("x+y = ",x + y)
elif c == "-":
    print("x-y = ",x - y)
elif c == "*":
    print("x*y = ",x * y)
elif c == "/":
    print("x/y = ",x / y)
elif c.lower() == "e":
    print("e^(x+y) = ",np.exp(x+y))
elif c.lower() == "sin":
    print("sin(x+y) = ",np.sin(x+y))
elif c.lower() == "cos":
    print("cos(x+y) = ",np.sin(x+y))
elif c.lower() == "pow":
    print("x^y = ",np.power(x,y))
```
